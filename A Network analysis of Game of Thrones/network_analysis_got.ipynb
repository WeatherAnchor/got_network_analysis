```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

s1_edges = pd.read_csv('data/got-s1-edges.csv')
s1_nodes = pd.read_csv('data/got-s1-nodes.csv')

s2_edges = pd.read_csv('data/got-s2-edges.csv')
s2_nodes = pd.read_csv('data/got-s2-nodes.csv')

s3_edges = pd.read_csv('data/got-s3-edges.csv')
s3_nodes = pd.read_csv('data/got-s3-nodes.csv')

s4_edges = pd.read_csv('data/got-s4-edges.csv')
s4_nodes = pd.read_csv('data/got-s4-nodes.csv')

s5_edges = pd.read_csv('data/got-s5-edges.csv')
s5_nodes = pd.read_csv('data/got-s5-nodes.csv')

s6_edges = pd.read_csv('data/got-s6-edges.csv')
s6_nodes = pd.read_csv('data/got-s6-nodes.csv')

s7_edges = pd.read_csv('data/got-s7-edges.csv')
s7_nodes = pd.read_csv('data/got-s7-nodes.csv')
G_book1 = nx.Graph()
G_book2 = nx.Graph()
G_book3 = nx.Graph()
G_book4 = nx.Graph()
G_book5 = nx.Graph()
G_book6 = nx.Graph()
G_book7 = nx.Graph()

screen_times = pd.read_csv('data/GOT_screentimes.csv')
```

```python
for _, edge in s1_edges.iterrows():
    G_book1.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'])
for _, node in s1_nodes.iterrows():
    G_book1.add_node(node['Id'], label=node['Label'])

for _, edge in s2_edges.iterrows():
    G_book2.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'])
for _, node in s2_nodes.iterrows():
    G_book2.add_node(node['Id'], label=node['Label'])
    
for _, edge in s3_edges.iterrows():
    G_book3.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'])
for _, node in s3_nodes.iterrows():
    G_book3.add_node(node['Id'], label=node['Label'])
    
for _, edge in s4_edges.iterrows():
    G_book4.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'])
for _, node in s4_nodes.iterrows():
    G_book4.add_node(node['Id'], label=node['Label'])
    
for _, edge in s5_edges.iterrows():
    G_book5.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'])
for _, node in s5_nodes.iterrows():
    G_book5.add_node(node['Id'], label=node['Label'])

for _, edge in s6_edges.iterrows():
    G_book6.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'])
for _, node in s6_nodes.iterrows():
    G_book6.add_node(node['Id'], label=node['Label'])

for _, edge in s7_edges.iterrows():
    G_book7.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'])
for _, node in s7_nodes.iterrows():
    G_book7.add_node(node['Id'], label=node['Label'])
```


* Degree centrality: how many people you know
* Weighted degree centrality: how many interactions you have
* Eigenvector centrality: how many important people you know
* PageRank: how many important interactions you have
* Betweenness: do you help to connect the network together?


### Хэр олон тооны хүмүүсийг таних вэ?
* Цувралын туршид хамгийн чухал хүн хэн болох, хэр олон тооны хүнтэй холбоо сүлбээтэй байдгийг мэдэхийн тулд "degree centrality" хэмээх алгоритмыг ашиглана. Энэ нь тухайн дүрийн танидаг ямар нэг байдлаар харилцаа тогтоосон дүрүүдийг мэдэж болох бүх дүрүүдтэй харицуулахад гарж ирэх юм

```python
books = [G_book1,G_book2,G_book3,G_book4,G_book5,G_book6,G_book7]
# Creating a list of degree centrality of all the books
evol = [nx.degree_centrality(book) for book in books]
# Creating a DataFrame from the list of degree centralities in all the books
evol_degc_df = pd.DataFrame(evol).fillna(0)
top10 = evol_degc_df.sort_values(axis=1, by=[6,5,4,3,2,1,0], ascending=False).iloc[:,0:10]
top10.plot(figsize=(13,7))
plt.show()
top10
```

### Хүрээлэлдээ амин судас буюу ихэнх мэдээлэл түүгээр дамжин өнгөрдөг ба өөрийн хүрээллээ сайн хадгалдаг дүрүүдийг олцгооё!
* Үүнийг олохын тулд "betweenness centrality" алгоритмыг ашиглах бөгөөд дээр дурдсан шиг алгоритмын үндсэн ажиллах зарчим нь хоёр талыг холбож буй дүр нь мэдээллийг хоёр талд хүргэж байгаа учир илүү их эрх мэдэлтэй болж байгаа юм.


```python
# Creating a list of betweenness centrality of all the books just like we did for degree centrality
bet_evol = [nx.betweenness_centrality(book, weight='Weight') for book in books]

# Making a DataFrame from the list
betweenness_evol_df = pd.DataFrame(bet_evol).fillna(0)
top10_betw = betweenness_evol_df.sort_values(axis=1, by=[6,5,4,3,2,1,0], ascending=False).iloc[:,0:10]
top10_betw.plot(figsize=(13,7))
plt.show()
top10_betw
```

### Хэр олон чухал уулзалт болон тохиолууд тухайн дүрд учирсан бэ?
* Үүнийг олохын тулд Google компаны хамгийн анхны хайлтын алгоритм болох "Pagerank"-г ашиглана
* Энэхүү алгоритм нь хүрээлэлд байх тухайн дүр нь хэдий олон удаа, хэр олон хүнтэй, хичнээн тооны чухал хүмүүстэй харилцаж өөрийн байр суурийг олсон бэ гэдгийг харуулахад туслах болно 
* Нэрнээс нь харахад л зорилго нь ойлгомжтой энэхүү алгоритмыг Google компани болон бусад компаниудад ашиглагддаг хайлтын илэрцүүдийг хамгийн чухалаас нь чухал бус хүртэл дараалалд оруулдан алгоритм юм

```python
pg_evol = [nx.pagerank(book) for book in books]
# Making a DataFrame from the list
pg_evol_df = pd.DataFrame(pg_evol).fillna(0)
top10_pg = pg_evol_df.sort_values(axis=1, by=[6,5,4,3,2,1,0], ascending=False).iloc[:,0:10]
top10_pg.plot(figsize=(13,7))
plt.show()
top10_pg
```

### Заримдаа тооноос илүү чанар чухал байдаг(quality > quantity)
* Олон тооны дүрүүдтэй холбоо тогтоосон, чухал хүмүүсийг таних нь заримдаа хэрэггүй ч байж мэднэ гэдгийг цуврал маань бидэнд сайн ойлгуулсныг санаж байгаа байх. Тиймээс энэ удаад дүрийн хэр олон хүмүүстэй дотно ба холбоо тогтооход ойрхон зэргээр дүгнэх "Closeness centrality" алгоритмыг ашиглах болно 
* Энэ нь тухайн дүрийг график дээр авч үзвэл бусад дүрүүдээс хамгийн багадаа хэр зайтай байгааг тооцоолоон нийлбэр гаргаж дүгнэлт гаргадаг алгоритм юм

```python
cc_evol = [nx.closeness_centrality(book) for book in books]
# Making a DataFrame from the list
cc_evol_df = pd.DataFrame(cc_evol).fillna(0)
top10_cc_evol = cc_evol_df.sort_values(axis=1, by=[6,5,4,3,2,1,0], ascending=False).iloc[:,0:10]
top10_cc_evol.plot(figsize=(13,7))
plt.show()
top10_cc_evol
```
